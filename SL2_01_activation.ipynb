```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def custom_sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
def custom_tanh(x):
  return np.tanh(x)
```

```python
def custom_reLu(x):
  return np.maximum(0, x)
```

```python
def custom_softmax(x):
  e_x = np.exp(x - np.max(x))
  return e_x / e_x.sum(axis=0)
```

```python
x = np.linspace(-5, 5, 100)
```

```python
sigmoid_fun = custom_sigmoid(x)
tanh_fun = custom_tanh(x)
reLu_fun = custom_reLu(x)
softmax_fun = custom_softmax(x)
```

```python
plt.plot(x, sigmoid_fun)
plt.title('Sigmoid Activation Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
```

```python
plt.plot(x, tanh_fun)
plt.title("tanh activation function")
plt.xlabel('x')
plt.ylabel('tanh_values')
```

```python
plt.plot(x, reLu_fun)
plt.title('reLu activation function')
plt.xlabel('x')
plt.ylabel('reLu_values')
```

```python
plt.plot(x, softmax_fun)
plt.title('softmax activation function')
plt.xlabel('x')
plt.ylabel('softmax_values')
```
